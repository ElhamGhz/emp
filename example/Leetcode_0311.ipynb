#### [Leetcode 311] [Sparse Matrix Multiplication](https://leetcode.com/problems/sparse-matrix-multiplication/)

Given two sparse matrices A and B, return the result of AB.

You may assume that A's column number is equal to B's row number.

Example:
```
Input:

A = [
  [ 1, 0, 0],
  [-1, 0, 3]
]

B = [
  [ 7, 0, 0 ],
  [ 0, 0, 0 ],
  [ 0, 0, 1 ]
]

Output:

     |  1 0 0 |   | 7 0 0 |   |  7 0 0 |
AB = | -1 0 3 | x | 0 0 0 | = | -7 0 3 |
                  | 0 0 1 |
```

````python
class Solution(object):
    def multiply1(self, A, B):
        """
        :type A: List[List[int]]
        :type B: List[List[int]]
        :rtype: List[List[int]]
        """
        if not A or not B:
            return []
        
        row_A, col_A = len(A), len(A[0])
        row_B, col_B = len(B), len(B[0])
        
        C = [[0 for _ in range(col_B)] for _ in range(row_A)]
        for i in range(row_A):
            for j in range(col_B):
                for k in range(col_A):
                    C[i][j] += A[i][k] * B[k][j]
                    
        return C
    
    def multiply2(self, A, B):
        """
        :type A: List[List[int]]
        :type B: List[List[int]]
        :rtype: List[List[int]]
        """
        if not A or not B:
            return []
        
        row_A, col_A = len(A), len(A[0])
        row_B, col_B = len(B), len(B[0])
        
        C = [[0 for _ in range(col_B)] for _ in range(row_A)]
        for i in range(row_A):
            for k in range(col_A):
                if A[i][k] != 0:
                    for j in range(col_B):
                        if B[k][j] != 0:
                            C[i][j] += A[i][k] * B[k][j]
                    
        return C
    
    def multiply3(self, A, B):
        """
        :type A: List[List[int]]
        :type B: List[List[int]]
        :rtype: List[List[int]]
        """
        if not A or not B:
            return []
        
        row_A, col_A = len(A), len(A[0])
        row_B, col_B = len(B), len(B[0])
        
        A_list = []
        for i in range(row_A):
            for j in range(col_A):
                if A[i][j] != 0:
                    A_list.append((i, j))
        
        B_list = []
        for i in range(row_B):
            for j in range(col_B):
                if B[i][j] != 0:
                    B_list.append((i, j))
        
        C = [[0 for _ in range(col_B)] for _ in range(row_A)]
        for A_i, A_j in A_list:
            for B_i, B_j in B_list:
                if A_j == B_i:
                    C[A_i][B_j] += A[A_i][A_j] * B[B_i][B_j]
                    
        return C
    
    def multiply(self, A, B):
        """
        :type A: List[List[int]]
        :type B: List[List[int]]
        :rtype: List[List[int]]
        """
        if not A or not B:
            return []
        
        row_A, col_A = len(A), len(A[0])
        row_B, col_B = len(B), len(B[0])
        
        
        B_hashmap = {}
        for k in range(row_B):
            for j in range(col_B):
                if B[k][j] != 0:
                    if k not in B_hashmap:
                        B_hashmap[k] = []
                    B_hashmap[k].append(j)
                    
        C = [[0 for _ in range(col_B)] for _ in range(row_A)]
        for i in range(row_A):
            for k in range(col_A):
                if A[i][k] != 0:
                    for j in B_hashmap[k]:
                        C[i][j] += A[i][k] * B[k][j]
           
        return C
````

````python
A = [
  [ 1, 0, 0],
  [-1, 0, 3]
]

B = [
  [ 7, 0, 0 ],
  [ 0, 0, 0 ],
  [ 0, 0, 1 ]
]
soln = Solution()
print(soln.multiply(A, B))
````
