#### [Leetcode 126 Hard] [Word Ladder II](https://leetcode.com/problems/word-ladder-ii/)

Given two words (beginWord and endWord), and a dictionary's word list, find all shortest transformation sequence(s) from beginWord to endWord, such that:
1. Only one letter can be changed at a time
2. Each transformed word must exist in the word list. Note that beginWord is not a transformed word.

Note:
* Return an empty list if there is no such transformation sequence.
* All words have the same length.
* All words contain only lowercase alphabetic characters.
* You may assume no duplicates in the word list.
* You may assume beginWord and endWord are non-empty and are not the same.

Example 1:
```
Input:
beginWord = "hit",
endWord = "cog",
wordList = ["hot","dot","dog","lot","log","cog"]

Output:
[
  ["hit","hot","dot","dog","cog"],
  ["hit","hot","lot","log","cog"]
]
```

Example 2:
```
Input:
beginWord = "hit"
endWord = "cog"
wordList = ["hot","dot","dog","lot","log"]

Output: []

Explanation: The endWord "cog" is not in wordList, therefore no possible transformation.
```

<font color='blue'>Solution: BFS + DFS </font>

````python
import string
from collections import defaultdict

class Solution(object):
    def findLadders(self, beginWord, endWord, wordList):
        """
        :type beginWord: str
        :type endWord: str
        :type wordList: List[str]
        :rtype: List[List[str]]
        """
        wordList = set(wordList)
        if endWord not in wordList:
            return []
        
        frontier = []
        frontier.append(beginWord)
        count = 1
        
        costs = {}
        costs[beginWord] = 1
        
        came_from = defaultdict(list)
        #came_from[beginWord] = []
        
        found = False
        
        while frontier and not found:
            post = []
            
            for curr_word in frontier:
                
                if curr_word == endWord:
                    found = True
                    break

                for i in range(len(curr_word)):
                    for char in string.ascii_lowercase:
                        next_word = curr_word[:i] + char + curr_word[i+1:]
                        
                        # Not a new word, but another transform
                        #   with the same number of steps
                        if next_word in costs and count + 1 <= costs[next_word]:
                            came_from[next_word].append(curr_word)

                        if next_word in wordList and next_word not in costs:
                            post.append(next_word)                                                       
                            came_from[next_word].append(curr_word)
                            costs[next_word] = costs[curr_word] + 1
            
            count += 1
            frontier = post
                        
        #print(came_from)
        
        results = []
        if found:
            curr = [endWord]
            # multi-stage decision-making problem
            self.dfs(results, curr, came_from, endWord, beginWord)
            
        return results
    
    def dfs(self, results, curr, came_from, word, beginWord):
        if word == beginWord:
            results.append(curr[::-1])
            return
        
        for parent in came_from[word]:
            curr.append(parent)
            self.dfs(results, curr, came_from, parent, beginWord)
            curr.pop()
    
if __name__ == "__main__":
    soln = Solution()

    beginWord = "hit"
    endWord = "cog"
    wordList = ["hot","dot","dog","lot","log","cog"]
    print(soln.findLadders(beginWord, endWord, wordList))

    beginWord = "hit"
    endWord = "cog"
    wordList = ["hot","dot","dog","lot","log"]
    print(soln.findLadders(beginWord, endWord, wordList))

    beginWord = "red"
    endWord = "tax"
    wordList = ["ted","tex","red","tax","tad","den","rex","pee"]
    print(soln.findLadders(beginWord, endWord, wordList))
````
