#### [Leetcode 445 Medium] [Add Two Numbers II](https://leetcode.com/problems/add-two-numbers-ii/) 

You are given two non-empty linked lists representing two non-negative integers. The most significant digit comes first and each of their nodes contain a single digit. Add the two numbers and return it as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

Follow up:
What if you cannot modify the input lists? In other words, reversing the lists is not allowed.

Example:
```
Input: (7 -> 2 -> 4 -> 3) + (5 -> 6 -> 4)
Output: 7 -> 8 -> 0 -> 7
```

````python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def addTwoNumbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        # Edge case
        if not l1:
            return l2
        if not l2:
            return l1
        
        # reverse
        head1 = self.reverse(l1)
        head2 = self.reverse(l2)
        #print('reverse finished')
        #linked_list = LinkedList()
        #linked_list.printList(head1)
        #linked_list.printList(head2)
        
        # addition
        carry = 0
        fake_head = ListNode(None)
        curr = fake_head
        while head1 and head2:
            temp = head1.val + head2.val + carry
            carry = temp // 10
            # create new connection
            curr.next = ListNode(temp % 10)
            # move forward
            curr = curr.next
            head1 = head1.next
            head2 = head2.next
            
        while head1:
            temp = head1.val + carry
            carry = temp // 10
            # create new connection
            curr.next = ListNode(temp % 10)
            # move forward
            curr = curr.next
            head1 = head1.next
        while head2:
            temp = head2.val + carry
            carry = temp // 10
            # create new connection
            curr.next = ListNode(temp % 10)
            # move forward
            curr = curr.next
            head2 = head2.next
            
        if carry:
            curr.next = ListNode(carry)
        
        # reverse
        head = self.reverse(fake_head.next)
        
        return head
        
    def reverse(self, head):
        # base case: -- tail node
        if not head or not head.next:
            return head
        
        # what to get from your child
        new_head = self.reverse(head.next)
        
        # what to do in the current stage
        # head --> post <-- ... <-- new_head
        post = head.next
        # head <-- post <-- ... <-- new_head
        # create new connection
        post.next = head
        # remove old connection
        head.next = None
        
        # what to return to your parent
        return new_head
        
    def reverse1(self, head):
        prev, curr = None, head
        while curr:
            # prev --> (curr) --> post
            post = curr.next
            # prev <-- (curr) --> post
            curr.next = prev
            # prev <-- curr --> (post)
            prev, curr = curr, post
        return prev
    
if __name__ == "__main__":
    linked_list = LinkedList()
    soln = Solution()
    
    # case 1
    head1 = ListNode(7)
    head1.next = ListNode(2)
    head1.next.next = ListNode(4)
    head1.next.next.next = ListNode(3)
    linked_list.printList(head1)
    
    head2 = ListNode(5)
    head2.next = ListNode(6)
    head2.next.next = ListNode(4)
    linked_list.printList(head2)
    
    
    head = soln.addTwoNumbers(head1, head2)
    linked_list.printList(head)
    
    # case 2:
    head1 = ListNode(0)
    linked_list.printList(head1)
    
    head2 = ListNode(7)
    head2.next = ListNode(3)
    linked_list.printList(head2)
    
    head = soln.addTwoNumbers(head1, head2)
    linked_list.printList(head)
    
````
