### 4.1.2 Tree Traversal
[Wiki_TreeTraversal](https://en.wikipedia.org/wiki/Tree_traversal)
In computer science, tree traversal (also known as tree search) is a form of graph traversal and refers to the process of visiting (checking and/or updating) each node in a tree data structure, exactly once. Such traversals are classified by the order in which the nodes are visited. The following algorithms are described for a binary tree, but they may be generalized to other trees as well.

* Depth-First Search (DFS): [wiki_DFS](https://en.wikipedia.org/wiki/Depth-first_search) if there are children nodes to explore, exploring them is definitely the priority. Depth-first search (DFS) is an algorithm for traversing or searching tree or graph data structures. The algorithm starts at the root node (selecting some arbitrary node as the root node in the case of a graph) and explores as far as possible along each branch before backtracking.
![DFS Example](source/lesson6_binarytree_DFS.png)

* Breadth-First Search (BFS): [wiki_BFS](https://en.wikipedia.org/wiki/Breadth-first_search) the priority is visiting every node on the same level we're currently on before visiting child nodes. Breadth-first search (BFS) is an algorithm for traversing or searching tree or graph data structures. It starts at the tree root (or some arbitrary node of a graph, sometimes referred to as a 'search key'), and explores all of the neighbor nodes at the present depth prior to moving on to the nodes at the next depth level.
![BFS Example](source/lesson6_binarytree_BFS.png)

#### Depth-First Search
These searches are referred to as depth-first search (DFS), as the search tree is deepened as much as possible on each child before going to the next sibling. For a binary tree, they are defined as display operations recursively at each node, starting with the root, whose algorithm is as follows:

The general recursive pattern for traversing a (non-empty) binary tree is this: At node N do the following:
* (L) Recursively traverse its left subtree. This step is finished at the node N again.
* (R) Recursively traverse its right subtree. This step is finished at the node N again.
* (N) Process N itself.

These steps can be done in any order. If (L) is done before (R), the process is called left-to-right traversal, otherwise it is called right-to-left traversal. The following methods show left-to-right traversal:
* Pre-Order Traversal: check off a node as you see it before you traverse any further in the tree
![pre-order traversal](source/lesson6_binarytree_DFS_preorder.png)
* In-Order Traversal: check off a node after you traverse its left child
![in-order traversal](source/lesson6_binarytree_DFS_inorder.png)
* Post-Order Traversal: check off a node only after you traverse all of its descendants
![post-order traversal](source/lesson6_binarytree_DFS_postorder.png)

#### Question 1.1: Pre-Order Traversal (先序[根]遍历)

*Solution 1*: Recursion

先处理自己，再左子节点，最后右子节点
```
                10 == root
               /  \
              5    15
             / \   / \
            2  7  12 20     <-- all leaf node's level == 3

10， 5， 2，7， 15， 12， 20
```

*Solution 2*: Iteration

```
                 A == root
               /   \
              B     C
             / \   / \ 
            N   N N   N     

第一次访问节点的时候打印
(A, 1) print --> (B, 1) print --> (A, 2) --> (C, 1) print
```

````python
class Traversal(object):
    def preorder_traverse(self, root):
        """
        # Time Complexity: O(n)
        # Space Complexity: O(h)  h - height of the binary tree
        """
        if not root:
            return []
        
        res = []
        self.helper(root, res)
        return res
        
    def helper(self, root, res):
        # Base Case
        if not root:
            return
        
        # Recursion Part
        res.append(root.val)
        self.helper(root.left, res)
        self.helper(root.right, res)
        
    def preorder_divideconquer(self, root):
        """
        # Time Complexity: O(n)
        # Space Complexity: O(h)  h - height of the binary tree
        """
        if not root:
            return []
        
        # what to get from yoour children -- divide
        left_res = self.preorder_divideconquer(root.left)
        right_res = self.preorder_divideconquer(root.right)
        
        # what to do in the current stage -- conquer
        results = []
        results.append(root.val)
        if left_res:
            results.extend(left_res)
        if right_res:
            results.extend(right_res)
        
        # what to return to your parent
        return results
        
    def preorder_iter(self, root):
        if not root:
            return []
        
        results = []
        stack = [(root, 1)]  # the first time to visit root node, (node, count)
        while stack:
            node, count = stack.pop()
            if count == 1:
                results.append(node.val)
                
                # for the visit of the second time (A, 2) -- > (C, 1)
                stack.append((node, count + 1))
                
                # add left node (B, 1)
                if node.left:
                    stack.append((node.left, 1))
                    
            elif count == 2:
                # add right node (C, 1)
                if node.right:
                    stack.append((node.right, 1))
                    
        return results
        
if __name__ == "__main__":
    root = TreeNode(10)
    root.left = TreeNode(5)
    root.right = TreeNode(15)
    root.left.left = TreeNode(2)
    root.left.right = TreeNode(7)
    root.right.left = TreeNode(12)
    root.right.right = TreeNode(20)

    soln = Traversal()
    print(soln.preorder_traverse(root))
    print(soln.preorder_divideconquer(root))
    print(soln.preorder_iter(root))
````

````python
class Traversal(object):
    def inorder_traverse(self, root):
        """
        # Time Complexity: O(n)
        # Space Complexity: O(h)  h - height of the binary tree
        """
        if not root:
            return []
        
        res = []
        self.helper(root, res)
        return res
        
    def helper(self, root, res):
        # Base Case
        if not root:
            return
        
        # Recursion Part        
        self.helper(root.left, res)
        res.append(root.val)
        self.helper(root.right, res)
        
    def inorder_divideconquer(self, root):
        # base case
        if not root:
            return []
        
        # what to get from your children -- divide
        left_res = self.inorder_divideconquer(root.left)
        right_res = self.inorder_divideconquer(root.right)
        
        # what to do in the current stage -- conquer
        results = []
        if left_res:
            results.extend(left_res)
        results.append(root.val)
        if right_res:
            results.extend(right_res)
        
        # what to return to your parent
        return results
        
    def inorder_iter(self, root):
        if not root:
            return []
        
        results = []
        stack = [(root, 1)]  # the first time to visit root node, (node, count)
        while stack:
            node, count = stack.pop()
            if count == 1:
                # for the visit of the second time (A, 2) -- > (C, 1)
                stack.append((node, count + 1))
                
                # add left node (B, 1)
                if node.left:
                    stack.append((node.left, 1))
                    
            elif count == 2:
                results.append(node.val)
                # add right node (C, 1)
                if node.right:
                    stack.append((node.right, 1))
                    
        return results
        
if __name__ == "__main__":
    root = TreeNode(10)
    root.left = TreeNode(5)
    root.right = TreeNode(15)
    root.left.left = TreeNode(2)
    root.left.right = TreeNode(7)
    root.right.left = TreeNode(12)
    root.right.right = TreeNode(20)

    soln = Traversal()
    print(soln.inorder_traverse(root))
    print(soln.inorder_divideconquer(root))
    print(soln.inorder_iter(root))
````

#### Question 1.3: Post-Order Traversal (后序[根]遍历)

*Solution 1*: Recursion

先左子节点，再右子节点，最后处理自己
```
                10 == root
               /  \
              5    15
             / \   / \
            2  7  12 20     <-- all leaf node's level == 3

2， 7， 5， 12， 20， 15， 10
```

*Solution 2*: Iteration

```
                 A == root
               /   \
              B     C
             / \   / \ 
            N   N N   N     

第三次访问节点的时候打印
(A, 1) --> (B, 1) --> N --> (B, 2) --> N --> (B, 3) print -->
--> (A, 2) --> (C, 1) --> N --> (C, 2)  --> N --> (C, 3) print --> 
--> (A, 3) print
```

````python
class Traversal(object):
    def postorder_traverse(self, root):
        """
        # Time Complexity: O(n)
        # Space Complexity: O(h)  h - height of the binary tree
        """
        if not root:
            return []
        
        res = []
        self.helper(root, res)
        return res
        
    def helper(self, root, res):
        # Base Case
        if not root:
            return
        
        # Recursion Part        
        self.helper(root.left, res)
        self.helper(root.right, res)
        res.append(root.val)
        
    def postorder_divideconquer(self, root):
        # base case
        if not root:
            return []
        
        # what to get from your children -- divide
        left_res = self.postorder_divideconquer(root.left)
        right_res = self.postorder_divideconquer(root.right)
        
        # what to do in the current stage -- conquer
        results = []
        if left_res:
            results.extend(left_res)        
        if right_res:
            results.extend(right_res)
        results.append(root.val)
        
        # what to return to your parent
        return results
    
    def postorder_iter(self, root):
        if not root:
            return []
        
        results = []
        stack = [(root, 1)]  # the first time to visit root node, (node, count)
        while stack:
            node, count = stack.pop()
            if count == 1:
                # for the visit of the second time (A, 2) -- > (C, 1)
                stack.append((node, count + 1))
                
                # add left node (B, 1)
                if node.left:
                    stack.append((node.left, 1))
                    
            elif count == 2:
                # for the visit of the third time (A, 3) -- > print
                stack.append((node, count + 1))
                
                # add right node (C, 1)
                if node.right:
                    stack.append((node.right, 1))
                    
            elif count == 3:
                results.append(node.val)
                    
        return results
        
if __name__ == "__main__":
    root = TreeNode(10)
    root.left = TreeNode(5)
    root.right = TreeNode(15)
    root.left.left = TreeNode(2)
    root.left.right = TreeNode(7)
    root.right.left = TreeNode(12)
    root.right.right = TreeNode(20)

    soln = Traversal()
    print(soln.postorder_traverse(root))
    print(soln.postorder_divideconquer(root))
    print(soln.postorder_iter(root))
````

#### Question 1.4: Level Order Traversal (BFS)

```
10
5, 15
2, 7, 12, 20
1
```

````python
class Traversal(object):
    def levelorder(self, root):
        """
        # Time Complexity: O(n)
        # Space Complexity: O(len(curr_level)+len(post_level)), O(n) at worst
        """
        if not root:
            return []
        
        results = []        
        
        frontier = [root]
        
        while frontier:
            post = []
            solution = [] # for a certain level
            for node in frontier:
                solution.append(node.val)
                if node.left:
                    post.append(node.left)
                if node.right:
                    post.append(node.right)
                    
            frontier = post
            results.append(solution)
            
        return results
    
    def BFS_traversal(self, root):
        # Time Complexity: O(n)
        # Space Complexity: O(len(curr_level)+len(post_level)), O(n) at worst
        if not root:  # if root is None
            return []
        
        results = [] # store the total results
        result_level = [] # store the result of the current level
        
        curr_level = [root]
        post_level = []
        
        while curr_level:  # if curr_level is not None
            head = curr_level.pop(0)
            if not head:
                result_level.append("#")
            else:                
                result_level.append(head.val)
                
                if head.left: # if head.left is not None
                    post_level.append(head.left)
                else:
                    post_level.append(None)
                
                if head.right: # if head.right is not None:
                    post_level.append(head.right)
                else:
                    post_level.append(None)            
            
            if not curr_level:  # if curr_level is None
                results.append(result_level)
                if post_level and self.not_all_none_check(post_level) :  # if post_level is not None
                    curr_level = post_level
                    post_level = []
                    result_level = []
                    
        return results
    
    def not_all_none_check(self, lst):
        if not lst:
            return False
        
        for val in lst:
            if val is not None:
                return True
            
        return False
    
if __name__ == "__main__":
    root = TreeNode(10)
    root.left = TreeNode(5)
    root.right = TreeNode(15)
    root.left.left = TreeNode(2)
    root.left.right = TreeNode(7)
    root.right.left = TreeNode(12)
    root.right.right = TreeNode(20)

    soln = Traversal()
    print(soln.levelorder(root))
    print(soln.BFS_traversal(root))
    
    root = TreeNode(10)
    root.left = TreeNode(5)
    root.right = TreeNode(15)
    #root.left.left = TreeNode(2)
    root.left.right = TreeNode(7)
    root.right.left = TreeNode(12)
    root.right.right = TreeNode(20)

    soln = Traversal()
    print(soln.levelorder(root))
    print(soln.BFS_traversal(root))
````

#### Question 1.5: [Leetcode 331] [Verify Preorder Serialization of a Binary Tree](https://leetcode.com/problems/verify-preorder-serialization-of-a-binary-tree/)

One way to serialize a binary tree is to use pre-order traversal. When we encounter a non-null node, we record the node's value. If it is a null node, we record using a sentinel value such as #.
```
     _9_
    /   \
   3     2
  / \   / \
 4   1  #  6
/ \ / \   / \
# # # #   # #
```
For example, the above binary tree can be serialized to the string "9,3,4,#,#,1,#,#,2,#,6,#,#", where # represents a null node.

Given a string of comma separated values, verify whether it is a correct preorder traversal serialization of a binary tree. Find an algorithm without reconstructing the tree.

Each comma separated value in the string must be either an integer or a character '#' representing null pointer.

You may assume that the input format is always valid, for example it could never contain two consecutive commas such as "1,,3".

Example 1:
```
Input: "9,3,4,#,#,1,#,#,2,#,6,#,#"
Output: true
```

Example 2:
```
Input: "1,#"
Output: false
```

Example 3:
```
Input: "9,#,#,1"
Output: false
```

[Solution](https://blog.csdn.net/fuxuemingzhu/article/details/79537797)

````python
class Solution(object):
    def isValidSerialization(self, preorder):
        """
        :type preorder: str
        :rtype: bool
        用一个栈，从字符串的左侧向右依次进栈，如果满足栈的后三位是数字，#，#的模式时，说明可以构成合法的叶子节点，
        把这个叶子节点换成#号，代表空节点，然后继续遍历。最后应该只剩下一个#，那么就是一个合法的二叉树。
        """
        stack = []
        for node in preorder.split(','):
            stack.append(node)
            while len(stack) >= 3 and stack[-1] == stack[-2] == '#' and stack[-3] != '#':
                stack.pop(), stack.pop(), stack.pop()
                stack.append('#')
        return len(stack) == 1 and stack.pop() == '#'

    def isValidSerialization(self, preorder):
        """
        :type preorder: str
        :rtype: bool
        我们知道一个树（甚至图），所有节点的入度之和等于出度之和。那么可以根据这个条件进行有效性的判断。
        对于二叉树，我们把空的地方也作为叶子节点（如题目中的#），那么有
          1. 所有的非空节点提供2个出度和1个入度（根除外）
          2. 所有的空节点但提供0个出度和1个入度
        我们在遍历的时候，计算diff = outdegree – indegree. 当一个节点出现的时候，diff – 1，因为它提供一个入度；
        当节点不是#的时候，diff+2(提供两个出度) 如果序列式合法的，那么遍历过程中diff >=0 且最后结果为0.
        
        这里解释一下为什么diff的初始化为1.因为，我们加入一个非空节点时，都会先减去一个入度，再加上两个出度。
        但是由于根节点没有父节点，所以其入度为0，出度为2.因此diff初始化为1，
        是为了再加入根节点的时候，先减去一个入度，再加上两个出度，正好应该是2.

        """
        nodes = preorder.split(',')
        diff = 1
        for node in nodes:
            diff -= 1
            if diff < 0:
                return False
            if node != '#':
                diff += 2
        return diff == 0
    
if __name__ == "__main__":
    soln = Solution()
    
    print(soln.isValidSerialization(preorder="9,3,4,#,#,1,#,#,2,#,6,#,#"))
    print(soln.isValidSerialization(preorder="1,#"))
    print(soln.isValidSerialization(preorder="9,#,#,1"))
````

#### Question 7.1: [Leetcode 549] Binary Tree Longest Consecutive Sequence II

Given a binary tree, find the length(number of nodes) of the longest consecutive sequence(Monotonic and adjacent node values differ by 1) path.

The path could be start and end at any node in the tree.

Especially, this path can be either increasing or decreasing. For example, [1,2,3,4] and [4,3,2,1] are both considered valid, but the path [1,2,4,3] is not valid. On the other hand, the path can be in the child-Parent-child order, where not necessarily be parent-child order.

Example 1:
```
Input:
{1,2,0,3}
Output:
4
Explanation:
    1
   / \
  2   0
 /
3
0-1-2-3
```

Example 2:
```
Input:
{3,2,2}
Output:
2
Explanation:
    3
   / \
  2   2
2-3
```

这道题是之前那道Binary Tree Longest Consecutive Sequence的拓展，那道题只让从父结点到子结点这种顺序来找最长连续序列，而这道题没有这个顺序限制，我们可以任意的拐弯，这样能找到最长的递增或者递减的路径。这道题利用回溯的思想比较容易，因为当一个结点没有子结点点时，它只需要跟其父结点进行比较，这种情况最容易处理，而且一旦叶结点处理完了，我们可以一层一层的回溯，直到回到根结点，然后再遍历的过程中不断的更新结果res即可。由于题目中说了要么是递增，要么是递减，我们不能一会递增一会递减，所以我们递增递减的情况都要统计，只是最后取最长的路径。所以我们要知道每一个结点的最长递增和递减路径的长度，当然是从叶结点算起，这样才方便往根结点回溯。当某个结点比其父结点值大1的话，说明这条路径是递增的，那么当我们知道其左右子结点各自的递增路径长度，那么当前结点的递增路径长度就是左右子结点递增路径长度中的较大值加上1，同理如果是递减路径，那么当前结点的递减路径长度就是左右子结点递减路径长度中的较大值加上1，通过这种方式我们可以更新每个结点的递增递减路径长度。在回溯的过程中，一旦我们知道了某个结点的左右子结点的最长递增递减路径长度，那么我们可以算出当前结点的最长连续序列的长度，要么是左子结点的递增路径跟右子结点的递减路径之和加1，要么事左子结点的递减路径跟右子结点的递增路径之和加1，二者中取较大值即可

````python
class Solution(object):
    def get_longest(self, root):
        """
        # Time Complexity: O(n)
        # Space Complexity: O(h)
        """
        self.longest_path = 0
        self.longest(root, None)
        return self.longest_path

    def longest(self, curr, parent):
        # Base Case
        if not curr:
            return 0, 0

        # What to get from your children
        left_inc, left_dec = self.longest(curr.left, curr)
        right_inc, right_dec = self.longest(curr.right, curr)
        
        # What to do in the current stage
        curr_len = max(left_inc + right_dec + 1, left_dec + right_inc + 1)
        if curr_len > self.longest_path:
            self.longest_path = curr_len

        # What to return to your parent
        if parent and curr.val + 1 == parent.val:
            curr_inc = max(left_inc, right_inc) + 1
        else:
            curr_inc = 0
        if parent and curr.val - 1 == parent.val:
            curr_dec = max(left_dec, right_dec) + 1
        else:
            curr_dec = 0
            
        return (curr_inc, curr_dec)

if __name__ == "__main__":
    root = TreeNode(1)
    root.left = TreeNode(2)
    root.right = TreeNode(0)
    root.left.left = TreeNode(3)

    soln = Solution()
    print(soln.get_longest(root))
````

````python
class Solution(object):
    def pathSum(self, root, target):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: List[List[int]]
        muti-stage problem -- backtracking
        """
        results = []
        if not root:
            return results
        
        solution = []
        self.backtracking(root, target, results, solution)
        return results
        
    def backtracking(self, root, cur_sum, results, solution):
        if not root:
            return
        if root.val == cur_sum and root.left is None and root.right is None:
            results.append(solution[:] + [root.val])
            return
        
        solution.append(root.val)
        self.backtracking(root.left, cur_sum - root.val, results, solution)
        solution.pop()
        
        solution.append(root.val)
        self.backtracking(root.right, cur_sum - root.val, results, solution)
        solution.pop()

if __name__ == "__main__":
    root = TreeNode(5)
    root.left = TreeNode(4)
    root.right = TreeNode(8)
    root.left.left = TreeNode(11)
    root.right.left = TreeNode(13)
    root.right.right = TreeNode(4)
    root.left.left.left = TreeNode(7)
    root.left.left.right = TreeNode(2)
    root.right.right.left = TreeNode(5)
    root.right.right.right = TreeNode(1)

    traverse = Traversal()
    print(traverse.BFS_traversal(root))

    soln = Solution()
    print(soln.pathSum(root, target=22))            
````

#### Question 10.2: [Leetcode 437] [Path Sum III](https://leetcode.com/problems/path-sum-iii/)

You are given a binary tree in which each node contains an integer value.

Find the number of paths that sum to a given value.

The path does not need to start or end at the root or a leaf, but it must go downwards (traveling only from parent nodes to child nodes).

The tree has no more than 1,000 nodes and the values are in the range -1,000,000 to 1,000,000.

Example:
```
root = [10,5,-3,3,2,null,11,3,-2,null,1], sum = 8

      10
     /  \
    5   -3
   / \    \
  3   2   11
 / \   \
3  -2   1

Return 3. The paths that sum to 8 are:

1.  5 -> 3
2.  5 -> 2 -> 1
3. -3 -> 11
```

*Solution*: Recursion

Like #112, we should recursively find matched path from root. However, it’s different that we have to do pathSum for left and right subtrees as well. Besides, once there is matched path found, it shouldn’t stop so the rest path will be examined, too.

If we assume counts of nodes in this tree is n, the recursion will run on $n + (n/2)*2 + (n/4)*4 + … + 1*n$ nodes. Therefore, it’s time complexity is $O(n\log{n})$.

For space complexity, the maximum calls in calling stack will be h if h denotes to the length of the deepest path in this tree. Thus, it uses $O(h)$ extra space. Or you can say $O(logn)$ because h is bounded by logn.

````python
class Solution(object):
    def pathSum(self, root, target):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: int
        """
        if not root:
            return 0
        
        # what to get from your children
        left_sum = self.pathSum(root.left, target)
        right_sum = self.pathSum(root.right, target)
        
        # what to do in the current stage
        curr_sum = self.helper(root, target)
        
        # what to return to your parent
        total_sum = curr_sum + left_sum + right_sum
        return total_sum
    
    def helper(self, root, curr_sum):
        # Base case
        if not root:
            return 0
        
        # what to do in the current stage
        curr_count = 0
        if root.val == curr_sum:
            curr_count = 1
            
        # what to get from your children
        left_count = self.helper(root.left, curr_sum - root.val)
        right_count = self.helper(root.right, curr_sum - root.val)
        
        # what to return to your parent
        total_count = curr_count + left_count + right_count
        return total_count
    
if __name__ == "__main__":
    root = TreeNode(10)
    root.left = TreeNode(5)
    root.right = TreeNode(-3)
    root.left.left = TreeNode(3)
    root.left.right = TreeNode(2)
    root.right.right = TreeNode(11)
    root.left.left.left = TreeNode(3)
    root.left.left.right = TreeNode(-2)
    root.left.right.right = TreeNode(1)

    traverse = Traversal()
    print(traverse.BFS_traversal(root))

    soln = Solution()
    print(soln.pathSum(root, target=8))     
````

*Solution*: — Hash Map

If we observe recursive behavior in recursion approach, it will be found that we always start with one node and traverse all it’s possible path downward. Therefore, if we can save accumulative sum from the each root node to current node in a hash map, a matched path can be easily looked up by hash_map.get(acc_sum — sum).

For example, assume there is a path 5 -> -4 -> 6 -> -1 and sum = 2. Because all accumulative sums are all in the hash map, which means it will be {5: 1, 1: 1, 7: 1, 6: 1}, then acc_sum — sum = [5 + (-4) + 6] — 2 = 5 should be in the hash map, which also means that -4 -> 6 is a path whose sum is equal to sum = 2.

By the idea mentioned above, we can pick matched count in a bottom-up manner so this approach can be very performant.

Because this approach will only visit each node once, it’s time complexity is O(n) if n denotes to counts of nodes in this tree. However, it may need to record accumulative sum for each node so it uses O(n) extra space.

````python
class Solution(object):
    def pathSum(self, root, target):
        """
        :type root: TreeNode
        :type sum: int
        :rtype: int
        """
        if not root:
            return 0
        
        self.counter = {}
        prefixsum = 0
        self.counter[prefixsum] = 1
        
        return self.helper(root, target, prefixsum)
    
    def helper(self, root, target, prefixsum):
        # Base case
        if not root:
            return 0
        
        # what to do in the current stage
        currsum = prefixsum + root.val
        self.counter[currsum] = self.counter.get(currsum, 0) + 1 
            
        # what to get from your children
        left_count = self.helper(root.left, target, currsum)
        right_count = self.helper(root.right, target, currsum)
        
        self.counter[currsum] -= 1
        
        # what to return to your parent
        curr_count = self.counter.get(currsum - target, 0)
        total_count = curr_count + left_count + right_count
        return total_count
    
if __name__ == "__main__":
    root = TreeNode(10)
    root.left = TreeNode(5)
    root.right = TreeNode(-3)
    root.left.left = TreeNode(3)
    root.left.right = TreeNode(2)
    root.right.right = TreeNode(11)
    root.left.left.left = TreeNode(3)
    root.left.left.right = TreeNode(-2)
    root.left.right.right = TreeNode(1)

    traverse = Traversal()
    print(traverse.BFS_traversal(root))

    soln = Solution()
    print(soln.pathSum(root, target=8))   
````

If the depth of a tree is smaller than 5, then this tree can be represented by a list of three-digits integers.

For each integer in this list:
The hundreds digit represents the depth D of this node, 1 <= D <= 4.
The tens digit represents the position P of this node in the level it belongs to, 1 <= P <= 8. The position is the same as that in a full binary tree.
The units digit represents the value V of this node, 0 <= V <= 9.
Given a list of ascending three-digits integers representing a binary with the depth smaller than 5. You need to return the sum of all paths from the root towards the leaves.

Example 1:
```

Output: 12

The tree that the list represents is:
    3
   / \
  5   1

```

Example 2:
```
Input: [113, 221]

Explanation: 
The tree that the list represents is: 
    3
     \
      1

The path sum is (3 + 1) = 4.
```

````python
class Solution:
    """
    @param nums: a list of integers
    @return: return an integer
    """

    def pathSum(self, nums):
        n, siz = len(nums), 1
        self.ans, self.mx = 0, nums[n - 1] // 100
        for i in range(self.mx - 1):
            siz *= 2
        g = [[-1 for j in range(siz)]for i in range(self.mx)]
        for i in range(n):
            dep, pos = nums[i] // 100, nums[i] // 10 % 10
            g[dep - 1][pos - 1] = nums[i] % 10
        self.dfs(g, 0, 0, 0)
        return self.ans

    def dfs(self, g, d, p, sum):
        if(g[d][p] == -1):
            return
        sum += g[d][p]
        if(d == self.mx - 1 or (g[d + 1][2 * p] == -1 and g[d + 1][2 * p + 1] == -1)):
            self.ans += sum
            return
        self.dfs(g, d + 1, 2 * p, sum)
        self.dfs(g, d + 1, 2 * p + 1, sum)
````

#### Question 11.1 (Follow Up): LCA II -- with Parent Node

Given the root and two nodes in a Binary Tree. Find the lowest common ancestor(LCA) of the two nodes.

The lowest common ancestor is the node with largest depth which is the ancestor of both nodes.

The node has an extra attribute parent which point to the father of itself. The root's parent is null.

Assume we have a new type 
```
class ParentTreeNode(object):
    def __init__(self, val):
        self.val = val
        self.parent = None
        self.left = None
        self.right = None
```
can you find the LCA with iteration method?

A和B往上总会在一个地方相交，这个地方就LCA。如果A和B不在同一个高度上，我们可以把一个点往上沿着parent指针跳到和另一个相同高度的位置。两个相同高度的点同时沿着各自的parent往上爬，爬到相交的地方就是LCA了。

````python
class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """
        # Edge case：
        if not root:
            return None
        elif root == p or root == q:
            return root
        
        depth_p = self.getDepth(p)
        depth_q = self.getDepth(q)
        
        if depth_q > depth_p:
            p, q = q, p
            
        diff = abs(depth_p - depth_q)
        while diff:
            p = p.parent
            diff -= 1
            
        while p != q:
            p = p.parent
            q = q.parent
            
        return p        
        
    def getDepth(self, node):
        depth = 0
        while node:
            depth += 1
            node = node.parent
            
        return depth
````

````python
class Solution(object):
    def isBalanced(self, root):
        """
        input: TreeNode root
        return: boolean
        """
        # write your solution here
        if not root:
            return True
        
        res, height = self.helper(root)
        return res
    
    def helper(self, root):
        # Base Case:
        if not root:
            return (True, 0)
        
        # What to get from your children
        left_res, left_height = self.helper(root.left)
        right_res, right_height = self.helper(root.right)
        
        # what to do in the current stage
        curr_height = max(left_height, right_height) + 1
        if not left_res or not right_res:
            curr_res = False
        elif abs(left_height - right_height) > 1:
            curr_res = False
        else:
            curr_res = True
        
        # What to return to your parent
        return (curr_res, curr_height)
    
if __name__ == "__main__":
    root = TreeNode(5)
    root.left = TreeNode(3)
    root.right = TreeNode(8)
    root.left.left = TreeNode(1)
    root.left.right = TreeNode(4)
    root.right.right = TreeNode(11)

    soln = Solution()
    print(soln.isBalanced(root))

    root = TreeNode(5)
    root.left = TreeNode(3)
    root.left.left = TreeNode(1)
    root.left.right = TreeNode(4)

    print(soln.isBalanced(root))
````

#### Question 21\*: Identical Binary Tree Structures

Let's assume if we tweak the left child with right child of an arbitrary node in a binary tree, then the structure of the tree are not changed. Then how can we determine whether two bineary trees structures are identical.

Example
```
                       8
                     /   \
                    5     4
case1. 8a         8b                 case2. 8a        8b
      /  \       /  \                      /  \      /  \
     4a  5a     4b  5b                    4a  5a    5b  4b
     |          |                         |             |
     7          7                         7             7
root1.left == root2.left && root1.right == root2.right
root1.left == root2.right && root1.right == root2.left
```

````python
class Solution(object):
    def helper(self, root1, root2):
        """
        Time Complexity: O(4^h) = O(4^(logn)) = O(2^(2logn)) = O(n^2)
        """
        # Base case
        if not root1 and not root2:
            return True
        elif not root1 or not root2:
            return False
        
        # What to do in the current stage
        if root1.val != root2.val:
            return False
        
        # What to get from your children
        res1 = self.helper(root1.left, root2.left)
        res2 = self.helper(root1.right, root2.right)
        res3 = self.helper(root1.left, root2.right)
        res4 = self.helper(root1.right, root2.left)
        
        # What to return to your parent
        return (res1 and res2) or (res3 and res4)
````

*Solution 3*:  

Way of thinking (Tricks)
1. What do you expect from your left child / right child? (usudally it is the return type of the recursion function)

 ```
                /    \         /           \

               \
(T, None, None) None
```

````python
class Solution(object):
    def is_BST_Top_Down(self, root): # 从上往下 From the Top Down 
        """
        把 上下界 往下传 Pass (lower_bound, upper_bound) from the top down
        1. How do you divide the problem into smaller ones?  
           Ans: 
          * my parents passes me a (lower_bound, upper_bound), root.value should be within this range
          * I pass (lower_bound, root.val) to my left child, it should fulfill this requirement
          * I pass (root.val, upper_bound) to my right child, it should fulfill this requirement
        2. What is the smallest problem -- base case condition?   
           Ans: root is None. (None under the leaf node)
        """
        if root is None:
            return True
        min_val = float('-inf')
        max_val = float('inf')
        return self.is_BST_Top_Down_helper(root, min_val, max_val)
    
    def is_BST_Top_Down_helper(self, root, min_val, max_val):
        # Base Case:
        if root is None:
            return True
        
        # What to do in the current stage
        if root.val <= min_val or root.val >= max_val:
            return False
        
        # What to get from my children
        left_res = self.is_BST_Top_Down_helper(root.left, min_val, root.val)
        right_res = self.is_BST_Top_Down_helper(root.right, root.val, max_val)
        
        # what to return to my parent
        curr_res = left_res and right_res
        return curr_res
        
    def is_BST_in_order(self, root): # 从上往下 Up --> Down
        # Edge Case
        if root is None:
            return True
        
        prev_val = [None]
        res = [True]
        self.in_order(root, prev_val, res)
        return res[0]
    
    def in_order(self, root, prev_val, res):
        # Base Case
        if not root:
            return
        
        self.in_order(root.left, prev_val, res)
        if prev_val[0] is not None and prev_val[0] >= root.val:
            res[0] = False
            
        prev_val[0] = root.val
        self.in_order(root.right, prev_val, res)
        
    def is_BST_Bottom_Up(self, root):
        """
        把 （判断结果，左子树最小值，右子树最大值） 往上传  from the bottom up
        1. How do you divide the problem into smaller ones?  
           Ans: 
          * its left subtree is a BST,
          * its right subtree is a BST, and
          * max_val in left subtree < root.val < min_val in right subtree
        2. What is the smallest problem -- base case condition?   
           Ans: root is None. (None under the leaf node)
                return (True, None, None)
        3. What to get from your children?
           Ans: 
          * left_child: (is BST, min_val in left_child, max_val in left_child)
          * right_child: (is BST, min_val in right_child, max_val in right_child)
        4. What to do in the current stage?
           Ans: 
          * its left subtree is a BST,
          * its right subtree is a BST, and
          * max_val in left subtree < root.val
          * root.val < min_val in right subtree
        5. What to return to your parent?
           Ans: (is BST, min_val in left_child or root.val, max_val in right_child or root.val)           
        """
        # Edge case
        if not root:
            return True
        
        res, min_val, max_val = self.is_BST_Bottom_Up_helper(root)
        return res
    
    def is_BST_Bottom_Up_helper(self, root):
        # Base Case:
        if not root:
            return (True, None, None)
        
        # What to get from your left and right children
        left_res, left_min, left_max = self.is_BST_Bottom_Up_helper(root.left)
        right_res, right_min, right_max = self.is_BST_Bottom_Up_helper(root.right)
        
        # What to do in the current stage
        if not left_res or not right_res:
            return (False, None, None)
        if left_max and left_max >= root.val:
            return (False, None, None)
        if right_min and right_min <= root.val:
            return (False, None, None)
        
        # What to return to your parent
        # return (True, left_min or root.val, right_max or root.val)
        curr_res = True
        curr_min = left_min or root.val
        curr_max = right_max or root.val
        return (curr_res, curr_min, curr_max)
    
if __name__ == "__main__":
    root = TreeNode(10)
    root.left = TreeNode(5)
    root.right = TreeNode(15)
    root.left.left = TreeNode(2)
    root.left.right = TreeNode(7)
    root.right.left = TreeNode(12)
    root.right.right = TreeNode(20)
    traversal = Traversal()
    print(traversal.BFS_traversal(root))

    soln = Solution()
    print(soln.is_BST_Top_Down(root))
    print(soln.is_BST_in_order(root))
    print(soln.is_BST_Bottom_Up(root))



    root = TreeNode(10)
    root.left = TreeNode(5)
    root.right = TreeNode(15)
    root.left.right = TreeNode(100)
    traversal = Traversal()
    print(traversal.BFS_traversal(root))

    soln = Solution()
    print(soln.is_BST_Top_Down(root))
    print(soln.is_BST_in_order(root))
    print(soln.is_BST_Bottom_Up(root))
````

### 4.2.1 Why Binary Search Tree?

#### Concept

A binary search tree is a binary tree where each node has a comparable key (optionally, an aoosciated value) taht has the following property: for every single node in this tree, all keys contained within its left subtree is smaller than the one in this single node, and all keys contained within its right subtree is larger than the one in this single node.

```
               /  \
              5    15
             / \   / \
            2  7  12 20  
```
If we do an in-order traversal on a binary search tree (recrusively traverse on the left subtree and then root node, then recursively traverse on the right subtree), what characteristics does the output sequence have?  
Ans: All items in this sequence will be organized in an increasing order.

Recall the concept "collection" that supports insert, remove and update items, previously, we have implemented this concept with queue, stack and single linked list. #### Implementation: Given a Key and an associated value, how do we insert them into the binary search tree?
```python
insert(root, key, value) := ... return the root node after the insertion
```
Recursion rule:  
```python

    return root
elif root.key < key:

    root.right = new_root_of_right_subtree

    new_root_of_left_subtree = insert(root.left, key, value)

    return root
```
Key observations: intuitively, it should be similar to what query does: if we find the key, then just update; otherwise, all that we need to do is to replace that link with a new node containing the key since a search for a key not in the tree ends at a null link. Similar to query, the recursion algorithm will look like:
1. If the tree is empty, we return a new node containing the key and value.
2. If key is less than the key at the root, we set the left link to the result of inserting the key into the left subtree; otherwise, we set the right link instead.

How will the API look when we try to implement the above recursive algorithm?  
Ans: As for return values, returning a link to a Node is needed since we can reflect changes to the tree by assigning the result to the link of the root node used as argument.

````python
class BinarySearchTree(object):
    def __init__(self):
        self.__root = None
        
    def __insert(self, root, key, value):
        # Base case
        if not root:
            return _Node(key, value)
        
        if root.key == key:
            root.value = value
        elif root.key < key:
            child_root = self.__insert(root.right, key, value)
            root.right = child_root
        else:
            child_root = self.__insert(root.left, key, value)
            root.left = child_root
            
        return root
        
    def insert(self, key, value):
        self.__root = self.__insert(self.__root, key, value)
````

Iteration Rule:  
Think about what this recursion will do. Basically, it is very similar to what we will do in query. When we hit the null link, we also need to know what node we previously encounter and the direction (left or right) so that we could set the the new node to the correct link:

````python
class BinarySearchTree(object):
    def __init__(self):
        self.__root = None
               
    def insert(self, key, value):
        if not self.__root:
            self.__root = _Node(key, value)
            return
        
        curr, prev, is_left = self.__root, None, None
        
        while curr:
            prev = curr
            if key < curr.key:
                is_left = True
                curr = curr.left
            elif key > curr.key:
                is_left = False
                curr = curr.right
            else:
                curr.value = value
                break
        
        if not curr:
            node = _Node(key, value)
            if is_left:
                prev.left = node
            else:
                prev.right = node
````

Based on the definition, starting from the root node, the leftmost node is the node that contains the smallest key. Can it have left child? No, if it has the left child, then that node contains key that is even smaller. Can it have right child? Yes, since all keys in the right subtree will be greater than the one in root node, this does not change the fact that the current key in root node is the smallest. The above tree is an example.

Hence, the easiest way to remove the leftmost node while preserving the binary search tree's properties is to replace this node with its right child. Similar to private insert API, we could define the return value to be the root node of the tree after deleting the node that contains the smallest key. With this, we could have the following code:

````python
class BinarySearchTree(object):
    def __init__(self):
        self.__root = None
        
    def __delete_min(self, root):
        if not root.left:
            return root.right
        
        root.left = self.__delete_min(root.left)
        
        return root
````

*Question 2*: 

Alternative:  
Now, how do we delete nodes? Well, for node that has only one child, by comparison, it should be the same as deleting the leftmost node from the tree. For node that has two children, we are left with two links, but only have a place in the parent node for only one of them. What should we do then? Can we utilize the fact that we could easily delete the node that contains the smallest key in a tree in some ways?

Assuming currently we are trying to delete node x that contains two children, after the deletion, we could delete the node that contains the smallest key in the right subtree and use it as the new root node!

Note that this is not the only way to delete a node that contains two children. An alternative is to put the left subtree of node x to be the left subtree of the leftmost node in the right subtree of node x. The resulting tree will still preserve the binary tree property. This has one downside though, as you would imagine, the height of the tree might increase by a lot. In our method, the height will never increase.
```
                y                          y
               /                          /
              x              ==>       min(c2)
             / \                        /  \ 
            c1  c2                     c1  remove_min(c2)
Outcome --> Does not increase depth
```

````python
class BinarySearchTree(object):
    def __init__(self):
        self.__root = None
        
    def __min(self, root):
        if not root.left:
            return root
        return self.__min(root.left)
    
    def __delete_min(self, root):
        if not root.left:
            return root.right
        
        root.left = self.__delete_min(root.left)
        
        return root
    
    def __delete(self, root, key):
        if not root:
            return None
        
        if key < root.key:
            root.left = self.__delete(root.left, key)
        elif key > root.key:
            root.right = self.__delete(root.right, key)
        else:
            if not root.right:
                return root.left
            if not root.left:
                return root.right
            t = root
            root = self.__min(t.right)
            root.right = self.__delete_min(t.right)
            root.left = t.left
        
        return root
    
    def delete(self, key):
        self.__root = self.__delete(self.__root, key)
````

#### Implementation: Additional Useful Operations (Homework)
Select/rank

For select, it means given an integer k (k>=0), I want to find a node in our binary search tree that has k nodes precede it. For example, when k = 0, the node that contains the smallest key is the one we want since there is no other node that will precede it.

For rank, it means the reverse operation of select: basically, given a key, return the number of keys that precede it in the tree.

Essentially, both operation require the knowledge of the size of the left subtree of a given node since it describes the number of keys that will precede the key in the given node. We could do a counting every time when we try to select/rank, but this is too slow. To boost the performance, we could actually cache this result (the size of a binary search tree) into out node directly, and, of course, we then need to remember to update this bit correctly whenever the structure of our binary search tree is changed.

````python
class _Node(object):  # similar to define a private class
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.left = None
        self.right = None
        
class BinarySearchTree(object):
    def __init__(self):
        # initialized the internal to become a valid empty binary search tree.
        self.__root = None
    
    
    def qurey(self, key):
        # if key is already presented in our tree, the associated value will be returned
        # otherwise, None will be returned
        return self.__query(self.__root, key)
    
    def __query(self, root, key):
        # Base Case
        if not root:
            return None
        
        if root.key == key:
            return root.value
        elif root.key < key: # target is in the right
            return self.__query(root.right, key)
        else:
            return self.__query(root.left, key)
            
    def qurey_iter(self, key):
        # if key is already presented in our tree, the associated value will be returned
        # otherwise, None will be returned
        curr = self.__root
        
        while curr:
            if curr.key == key:
                return curr.value
            elif curr.key < key:
                curr = curr.right
            else:
                curr = curr.left
                
        return None
                
        
    def insert(self, key, value):
        # if key is already presented in our tree, the associated value will be updated by our input value
        # otherwise, a new key value pair will be put into our tree
        self.__root = self.__insert(self.__root, key, value)
    
    def __insert(self, root, key, value):
        # return the root node after insertion
        if not root:
            return _Node(key, value)
        
        if root.key == key:
            root.value = value
        elif root.key < key:
            new_root = self.__insert(root.right, key, value)
            # create connection
            root.right = new_root
        else:
            new_root = self.__insert(root.left, key, value)
            root.left = new_root
            
        return root
    
    def insert_iter(self, key, value):
        if not self.__root:
            self.__root = _Node(key, value)
            return
        
        curr, prev, is_left = self.__root, None, None
        
        while curr:
            prev = curr
            if key < curr.key:
                is_left = True
                curr = curr.left
            elif key > curr.key:
                is_left = False
                curr = curr.right
            else:
                curr.value = value
                break
        
        if not curr:
            node = _Node(key, value)
            if is_left:
                prev.left = node
            else:
                prev.right = node
    
    def delete(self, key):
        # if key is already presented in our tree, this key value pair will be removed
        # otherwise, this function is a no-op.
        self.__root = self.__delete(self.__root, key)
    
    def __min(self, root):
        if not root.left:
            return root
        return self.__min(root.left)
    
    def __delete_min(self, root):
        # return the root node after deletion
        if not root.left:
            return root.right
        
        new_root = self.__delete_min(root.left)
        root.left = new_root
        
        return root
    
    def __delete(self, root, key):
        if not root:
            return None
        
        if key < root.key:
            root.left = self.__delete(root.left, key)
        elif key > root.key:
            root.right = self.__delete(root.right, key)
        else:
            # only has a left child, return right child
            if not root.right:
                return root.left
            # only has a right child, return left child
            if not root.left:
                return root.right
            # has left and right child
            #      t (old_root, to be deleted)
            #    /  \
            #  c1   c2
            
            t = root
            root = self.__min(t.right)
            root.right = self.__delete_min(t.right)
            root.left = t.left
        
        return root
````

````python
class Solution(object):
    def search_BST(self, root, val):
        """
        # Basic solution
        # Time Complexity: O(n)
        :type root: TreeNode
        :type val: int
        :rtype: TreeNode
        """
        # Edge Case
        if root is None:
            return None
        # Base Case
        if root.val == val:
            return root
        
        # What do you expect from your left/right child
        left = self.search_BST(root.left, val)
        right = self.search_BST(root.right, val)
        
        # What do you want to do for the current layer
        # What do you want to return to your parent
        if left is not None:
            return left
        if right is not None:
            return right
        return None
    
    def search_BST(self, root, val):
        """
        # Use BST property
        # Time Complexity: O(logn), worst case (skew BST) O(n)
        :type root: TreeNode
        :type val: int
        :rtype: TreeNode
        """
        # Base Case
        if root is None:
            return None
        
        if root.val == val:
            return root
        elif root.val < val:
            return self.search_BST(root.right, val)
        else:
            return self.search_BST(root.left, val)

if __name__ == "__main__":
    root = TreeNode(4)
    root.left = TreeNode(2)
    root.left.right = TreeNode(1)
    root.left.right.right = TreeNode(3)
    root.right = TreeNode(7)
    trav = Traversal()
    print(trav.BFS_traversal(root))

    soln = Solution()
    print(soln.search_BST(root, val=2).val)
````

#### Question 4: Given the pre-order access sequence of a BST, output the post-order of the node. Suppose the nodes are all distinct in the labels.
Example
```
              10
            /    \
           5      15
         /  \    /  \  
        3   8   12   16
           / \
          6   9
Input:  [10, 5, 3, 8, 6, 9, 15, 12, 16]
Output: [3, 6, 9, 8, 5, 12, 16, 15, 10]
```

*Solution*:  
The first element A[0] in input is the root (pre-order). How do we know which elements belong to the left subtree and which to the right subtree?  
Start from the 2nd element until index k, such that A[k] > A[0]. Thus, A[1:k] belongs to the left subtree, and A[k:end] to the right subtree.

````python
class Solution(object):
    def pre_to_post(self, pre):
        """
        # Time Complexity: O(n)
        """
        if not pre:
            return []

        return self.post(0, len(pre)-1, pre)

    def post(self, low, high, pre):
        tree = []
        # Base case
        if low > high:
            return tree

        # what to get from your children
        bound = low + 1
        while bound <= high and pre[bound] < pre[low]:
            bound += 1

        left = self.post(low + 1, bound - 1, pre)
        right = self.post(bound, high, pre)
        
        # what to do in the current stage
        tree.extend(left)
        tree.extend(right)
        tree.append(pre[low])
        
        # what to return to your parent
        return tree

if __name__ == "__main__":
    soln = Solution()
    print(soln.pre_to_post(pre=[10, 5, 3, 8, 6, 9, 15, 12, 16]))
````

```
[1, 3, 4, 2], False

```

````python
class Solution(object):
    def isBSTPreorder(self, array):
        lowerbound = None
        stack = []
        for x in array:
            if lowerbound is not None and x < lowerbound: 
                return False
            while stack and stack[-1] < x: 
                lowerbound = stack.pop()
            stack.append(x)
        return True
    
if __name__ == "__main__":
    soln = Solution()
    
    print(soln.isBSTPreorder(array=[1, 2, 3]))
    print(soln.isBSTPreorder(array=[2, 1, 3]))
    print(soln.isBSTPreorder(array=[3, 2, 1, 5, 4, 6]))
    print(soln.isBSTPreorder(array=[1, 3, 4, 2]))
    print(soln.isBSTPreorder(array=[3, 4, 5, 1, 2]))
````

````python
class SegmentTreeNode(object):
    def __init__(self, start, end):
        # index start
        self.start = start
        # index end
        self.end = end
        # sum of array[start:end]
        self.sum = 0
        # left child node
        self.left = None
        # right child node
        self.right = None
````

````python
class NumArray(object):

    def __init__(self, nums):
        """
        :type nums: List[int]
        """
        self.array = nums
        self.root = self._build_segment_tree(nums, 0, len(nums) - 1)        

    def update(self, i, val):
        """
        :type i: int
        :type val: int
        :rtype: None
        """
        diff = val - self.array[i]
        # update array
        self.array[i] = val
        # update segement tree
        curr_node = self.root
        while curr_node:
            curr_node.sum += diff
            mid = curr_node.start + (curr_node.end - curr_node.start) // 2
            if i < mid:
                curr_node = curr_node.left
            else:
                curr_node = curr_node.right       

    def sumRange(self, i, j):
        """
        :type i: int
        :type j: int
        :rtype: int
        """
        return self._get_sum_from_tree(self.root, i, j)
    
    def _build_segment_tree(self, array, start, end):
        if start > end:
            return None
        curr_node = SegmentTreeNode(start, end)
        
        # base case: single number
        if start == end:
            curr_node.sum = array[start]
            return curr_node
        
        # what to get from your children
        mid = start + (end - start) // 2
        left_root = self._build_segment_tree(array, start, mid)
        right_root = self._build_segment_tree(array, mid + 1, end)
        
        # what to do in the current layer
        curr_node.left = left_root
        curr_node.right = right_root
        if curr_node.left:
            curr_node.sum += curr_node.left.sum
        if curr_node.right:
            curr_node.sum += curr_node.right.sum
            
        # what to ruturn to your parent
        return curr_node
    
    def _get_sum_from_tree(self, curr_node, start, end):
        # ( ) represent curr_node, __ represent [start:end]
        # __ ( ), ( ) __       
        if curr_node is None or curr_node.start > end or curr_node.end < start:
            return 0
        # _( )_
        elif curr_node.start >= start and curr_node.end <= end:
            return curr_node.sum
        # ( __ ), _( _ ), ( _ )_
        else:
            return self._get_sum_from_tree(curr_node.left, start, end) + self._get_sum_from_tree(curr_node.right, start, end)
        
        
# Your NumArray object will be instantiated and called as such:
# obj = NumArray(nums)
# obj.update(i,val)
# param_2 = obj.sumRange(i,j)

if __name__ == "__main__":
    soln = NumArray(nums=[2, 4, 4, 5, 6])
    soln.update(1, 5)
    param_2 = soln.sumRange(0,3)
    print(param_2)
````

### 4.3.3 Advanced Tree -- Trie Tree
Applications: search engine, for example search by prefix

Example
```
words: ab, abd, abch, cg, ef, ca

# method 1: use a dictionary to store words
dict = {ab, abd, abch, cg, ef, ca}

# method 2: use a trie tree to store words

                            root
             /               |               \     (maximum 26 children)
            a                c                e
            |b             /a  \g             |f  
           ab(T)        ca(T)  cg(T)          ef(T)    (T:end)
         /c     \d
        abc     abd(T)
        |h
      abch(T)

search(abch): return True
```

advantages: space
1. 如果是常规存储，比如存储在文本文件， space cost 是线性增加 O(nm), n 是单词的平均长度， m 是单词的个数
2. 但是，如果用 trie tree， space cost 是可以预估控制在一定数量级的 O(26^h), h 是单词的最大长度

````python
class TrieTreeNode(object):
    def __init__(self):
        self.children = [None] * 26 # or self.children = {}
        # indicate if it is the end of a word
        self.is_end = False
        # indicate how many words used this node
        self.count = 0
        
class TrieTree(object):
    def __init__(self):
        self.root = TrieTreeNode()
        
    def search(self, word):
        curr_node = self.root
        for i in range(len(word)):
            index = ord(word[i]) - ord('a')
            if curr_node.children[index] is None:
                return False
            else:
                curr_node = curr_node.children[index]
        return curr_node.is_end
    
    def add(self, word):
        curr_node = self.root
        for i in range(0, len(word)):
            index = ord(word[i]) - ord('a')
            if curr_node.children[index] is None:
                curr_node.children[index] = TrieTreeNode()
            curr_node.count += 1
            curr_node = curr_node.children[index]
        curr_node.is_end = True
        
    def delete(self, word):
        self._delete_helper(self.root, word, 0)
        
    def _delete_helper(self, curr_node, word, i):
        if i == len(word):
            curr_node.is_end = False
            return
        
        index = ord(word[i]) - ord('a')
        self._delete_helper(curr_node.children[index], word, i + 1)
        
        if curr_node.children[index].count == 0:
            curr_node.children[index] = None
        curr_node.count -= 1
        
if __name__ == "__main__":
    trie_tree = TrieTree()
    trie_tree.add("abc")
    print(trie_tree.search("abc"))

    trie_tree.add("abd")
    print(trie_tree.search("abc"))
    print(trie_tree.search("abd"))

    trie_tree.delete("abd")
    print(trie_tree.search("abc"))
    print(trie_tree.search("abd"))
````
